```python
# This is a Gradient Descent implementation in Python
```

```python
# This is our function we want to minimize
def fun(x, y):
    return  x**4 + (y - 3)**2 + 1
```

```python
# Just to be sure our function is well defined
fun(0,1)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Obtaining the contour plot (level curves)
x = np.linspace(-5,5, 50)
y = np.linspace(-1,8, 50)
X,Y = np.meshgrid(x,y)

z = fun(X,Y)
plt.figure()
levels = [0.0, 1.0, 2.0, 3.0, 7, 9, 10.0, 20.0]
cp = plt.contour(x,y,z, levels)
plt.clabel(cp, inline = True, fontsize = 10, fmt = '%2.1f')
plt.title('Contour Plot of fun(x,y)')
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.show()
```

```python
# This is the gradient of our function
def gradf(x):
    return [4*(x[0]**3), 2*(x[1]-3)]


# Initial values
x = np.array([0, 0])
alpha = 0.1

# Main part of the gradient decsent
y = x - alpha * np.array(gradf(x))
c = np.linalg.norm(y - x)

while c > 0.0001:
    x = y
    y = x - alpha * np.array(gradf(x))
    c = np.linalg.norm(y - x)

    
# Checking the results
print(x)
print(fun(x[0],x[1]))

    
```
