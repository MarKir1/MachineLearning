```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import scipy.stats as stats
from IPython.display import Image
import seaborn as sns
from IPython.core.pylabtools import figsize
```

```python
samples_all = []
samples_all_exp = []
samples_all_poisson = []
samples_all_geometric = []
samples_all_weibull = []
mu = .9
lam = 1.0
size=1000
n = 20000

for number_in_sample in range(1,n):
    samples = np.random.binomial(1, mu, size=size)
    samples_all.append(samples.mean())
    samples = np.random.exponential(scale=2.0,size=size)
    samples_all_exp.append(samples.mean())
    samples = np.random.geometric(p=.5, size=size)
    samples_all_geometric.append(samples.mean())
    samples = np.random.poisson (lam=lam, size=size)
    samples_all_poisson.append(samples.mean())
    samples = np.random.weibull(2.0, size=size)
    samples_all_weibull.append(samples.mean())

df = pd.DataFrame({ 'binomial' : samples_all, 
                     'poisson' : samples_all_poisson,
                     'geometric' : samples_all_geometric,
                    'exponential' : samples_all_exp,
                  'weibull' : samples_all_weibull})

from IPython.core.pylabtools import figsize
from matplotlib import pyplot as plt
figsize(17, 8)
fig, axes = plt.subplots(nrows=3, ncols=2)
df.binomial.hist(color='blue',ax=axes[0,0], alpha=0.9, bins=1000)
df.exponential.hist(ax=axes[0,1],color='r',bins=1000)
df.poisson.hist(ax=axes[1,0],color='r',bins=1000)
df.geometric.hist(ax=axes[1,1],color='r',bins=1000)
df.weibull.hist(ax=axes[2,0],color='r',bins=1000)

axes[0,0].set_title('Binomial')
axes[0,1].set_title('Poisson')
axes[1,0].set_title('Geometric')
axes[1,1].set_title('Exponential')
axes[2,0].set_title('Weibull')
```

#### Discrete Uniform Distribution

```python
n = 10000
results = []
results1 = []
for num_x in range(1,n):
    x = np.random.randint(low=1,high=7, size=num_x)
    mean_of_x = x.mean()
    results.append(mean_of_x)
    results1.append(mean_of_x - 3.5)

df = pd.DataFrame({ 'x' : results})

plt.figure(figsize=(10, 4))
sns.distplot(results1,kde=True)
plt.show()

figsize(11, 9)
df.plot(title='Law of Large Numbers Explained Shown Graphically',color='r')
plt.xlabel("Number of throws in sample")
plt.ylabel("Average Of Sample")
```

#### Exponential Distribution

```python
n = 10000
mu = .9
results = []
for num_x in range(1,n):
    x = np.random.exponential(scale=2.0,size=num_x)
    mean_of_x = x.mean()
    results.append(mean_of_x - 2.0)

df = pd.DataFrame({ 'x' : results})

plt.figure(figsize=(10, 4))
sns.distplot(results,kde=True)
plt.show()

figsize(11, 9)
df.plot(title='Law of Large Numbers Explained Shown Graphically',color='r')
plt.xlabel("Number of throws in sample")
plt.ylabel("Average Of Sample")
```

#### Geometric Distribution

```python
n = 10000
mu = .9
results = []
for num_x in range(1,n):
    x = np.random.geometric(p=.5, size=num_x)
    mean_of_x = x.mean()
    results.append(mean_of_x - 2)

df = pd.DataFrame({ 'x' : results})

plt.figure(figsize=(10, 4))
sns.distplot(results,kde=True)
plt.show()

figsize(11, 9)
df.plot(title='Law of Large Numbers Explained Shown Graphically',color='r')
plt.xlabel("Number of throws in sample")
plt.ylabel("Average Of Sample")
```

#### Poisson Distribution

```python
n = 10000
mu = .9
lam = 1.0
results = []
for num_x in range(1,n):
    x = np.random.poisson(lam=lam, size=num_x)
    mean_of_x = x.mean()
    results.append(mean_of_x - 1)

df = pd.DataFrame({ 'x' : results})

plt.figure(figsize=(10, 4))
sns.distplot(results,kde=True)
plt.show()

figsize(11, 9)
df.plot(title='Law of Large Numbers Explained Shown Graphically',color='r')
plt.xlabel("Number of throws in sample")
plt.ylabel("Average Of Sample")
```

#### Chi-squared Distribution

```python
n = 10000
results = []
for num_x in range(1,n):
    x = np.random.chisquare(4, size=num_x)
    mean_of_x = x.mean()
    results.append(mean_of_x - 4)

df = pd.DataFrame({ 'x' : results})

plt.figure(figsize=(10, 4))
sns.distplot(results,kde=True)
plt.show()

figsize(11, 9)
df.plot(title='Law of Large Numbers Explained Shown Graphically',color='r')
plt.xlabel("Number of throws in sample")
plt.ylabel("Average Of Sample")
```
